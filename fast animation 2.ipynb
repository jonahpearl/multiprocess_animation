```python
import numpy as np
import serial
import datetime as dt
import os
import matplotlib.pyplot as plt
import matplotlib
import time
from multiprocessing import Process, Queue
```

```python
from fast_animate import *

# Open queue to animator
data_queue = Queue()
n_samples = 400

# Start animator process. NB, this must be done via a function imported from a separate module/file due to quirks of multiprocessing + ipynb
animate_process = main_from_ipynb(data_queue, n_samples)

# Mimic data collection
for j in range(100):
    data = np.random.random((n_samples,))
#     t1 = time.time()
    data_queue.put(data)
#     print(time.time() - t1)  # report time to pickle this np array

# When finished with data collection, send empty tuple to indicate process should end
data_queue.put(tuple())

# Wait for process to finish
animate_process.join()
```
